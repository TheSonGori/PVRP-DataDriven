# 01 — Exploración visual de instancias del PVRP

Este notebook verifica visualmente la correcta carga de instancias del NEO Research Group y produce las figuras descriptivas que se incluirán en los Capítulos 1 y 4 de la memoria.

**Objetivos:**

1. Confirmar visualmente que las coordenadas, demandas y patrones se interpretan correctamente.
2. Comparar el tamaño y la estructura geométrica de las tres instancias de referencia (`p01`, `p23`, `pr06`).
3. Visualizar la **Best Known Solution** de `p01` para anclar la intuición sobre cómo luce una solución óptima del PVRP.

**Convención del repositorio:** todas las funciones de plotteo viven en `src/utils/visualization.py`. Este notebook solo las invoca.

```python
# Imports y configuración global
import sys
from pathlib import Path

# Permite que el notebook encuentre la carpeta `src` aunque se ejecute desde `notebooks/`
PROJECT_ROOT = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()
if str(PROJECT_ROOT) not in sys.path:
    sys.path.insert(0, str(PROJECT_ROOT))

import matplotlib.pyplot as plt

from src.data.instance_loader import load_instance
from src.data.solution_loader import load_solution
from src.utils.visualization import (
    plot_instance,
    plot_demand_distribution,
    plot_frequency_distribution,
    plot_solution,
)

DATA_DIR = PROJECT_ROOT / "data" / "raw"

# Estilo general de los gráficos
plt.rcParams["figure.dpi"] = 110
plt.rcParams["font.size"] = 10
```

## 1. Carga de las tres instancias de referencia

Las instancias representan tres escalas distintas:

| Instancia | Clientes | Horizonte | Vehículos | Capacidad | Comentario |
|-----------|----------|-----------|-----------|-----------|------------|
| `p01`     | 50       | 2 días    | 3         | 160       | Tamaño pequeño, frecuencia uniforme |
| `p23`     | 168      | 4 días    | 6         | 40        | Tamaño medio, distribución radial, frecuencias variadas |
| `pr06`    | 288      | 4 días    | 12        | 175       | Tamaño grande, configuración industrial |

```python
instances = {
    name: load_instance(DATA_DIR / f"{name}.txt")
    for name in ["p01", "p23", "pr06"]
}

for name, inst in instances.items():
    print(f"{name}: {inst}")
```

## 2. Disposición geográfica de los clientes

Cada figura muestra el depósito (cuadrado rojo) y los clientes (puntos). El color codifica la frecuencia de visita requerida en el horizonte de planificación.

```python
for name, inst in instances.items():
    fig = plot_instance(inst, color_by="frequency")
    plt.show()
```

### Observaciones

- **`p01`**: distribución aproximadamente uniforme alrededor del depósito; todos los clientes tienen frecuencia 1 (una sola visita en el horizonte de 2 días). Es un buen caso de prueba inicial porque elimina la dimensión "asignación de patrones".
- **`p23`**: estructura **radial concéntrica** con cinco niveles de distancia desde el origen. Los clientes interiores tienen frecuencia 4 (visita diaria), los intermedios frecuencia 2 y los exteriores frecuencia 1. Esta instancia permite verificar el manejo de patrones múltiples.
- **`pr06`**: distribución **dispersa e irregular**, más realista de un contexto industrial. La densidad varía por zona, y la mezcla de frecuencias (1, 2 y 4) requiere decisiones temporales no triviales.

## 3. Distribución de demandas

Verificación de que ninguna demanda individual excede la capacidad vehicular (condición necesaria para factibilidad).

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, (name, inst) in zip(axes, instances.items()):
    plot_demand_distribution(inst, ax=ax)
plt.tight_layout()
plt.show()
```

## 4. Distribución de frecuencias de visita

Esta es la dimensión que diferencia al PVRP del VRP clásico: cuántos clientes deben visitarse cuántas veces.

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 5))
for ax, (name, inst) in zip(axes, instances.items()):
    plot_frequency_distribution(inst, ax=ax)
plt.tight_layout()
plt.show()
```

## 5. Visualización de la Best Known Solution de `p01`

Cargamos la solución de referencia publicada por el NEO Research Group y la dibujamos. Cada subplot muestra un día del horizonte; las rutas se diferencian por color.

Esta figura sirve como **referencia visual** para identificar más adelante si las soluciones generadas por nuestros métodos (Greedy, VNS, RL) son razonables.

```python
p01 = instances["p01"]
bks = load_solution(DATA_DIR / "p01.res")

fig = plot_solution(p01, bks)
plt.show()
```

## 6. Resumen estadístico de la BKS

El método `summary()` produce las métricas que utilizaremos sistemáticamente en el Capítulo 4 para comparar los distintos métodos de resolución.

```python
summary = bks.summary(p01, bks_cost=bks.reported_cost)

print("Resumen de la Best Known Solution sobre p01")
print("-" * 50)
for key, value in summary.items():
    if isinstance(value, float):
        print(f"  {key:25s} = {value:.4f}")
    else:
        print(f"  {key:25s} = {value}")
```

## 7. Conclusión del análisis exploratorio

**Validaciones logradas:**

1. ✓ Las tres instancias se cargan correctamente y producen disposiciones geográficas coherentes.
2. ✓ Las demandas individuales nunca exceden la capacidad vehicular.
3. ✓ Las frecuencias y patrones se decodifican consistentemente con la documentación del NEO.
4. ✓ La Best Known Solution de `p01` se visualiza correctamente y reproduce el costo reportado (524.61).

**Implicaciones para los siguientes pasos:**

- `p01` será la instancia de **desarrollo y debugging** para los métodos Greedy, VNS y el agente de RL.
- `p23` permitirá validar el manejo correcto de patrones de visita múltiples.
- `pr06` se reservará como **instancia de evaluación final** para medir escalabilidad de los métodos en el Capítulo 4.
